```python
""" PACK GEOMETRY MANAGER:... """

mainwindow = tkinter.Tk()

mainwindow.title('Hello world')
mainwindow.geometry('640x480+8-400') # -/+ to minimize or maximize the sizes of widgets of window.


## label, canvas, are widgets

label = tkinter.Label(mainwindow, text='Hello World')
label.pack(side='top') # using pack(method) to fit the label of window.

leftframe = tkinter.Frame(mainwindow)
leftframe.pack(side='left', fill=tkinter.Y, expand=False)

canvas = tkinter.Canvas(leftframe, relief='raised', borderwidth=1) # instead of mainwindow this time leftframe bcoz we placed, mainwindow inside leftframe
#canvas = tkinter.Canvas(mainwindow, relief='raised', borderwidth=1)
## check with this below codes for vertical, horizontal expansions 
#canvas.pack(side='left', fill=tkinter.Y) ## fill used to fill the widget(window) '*vertically(Y) or hortizonally(X) or BOTH*'
#canvas.pack(side='left', fill=tkinter.X, expand=True) # using expand=True will work the 'X' OR 
#canvas.pack(side='top', fill=tkinter.X) ## use side='top'
#canvas.pack(side='left', fill=tkinter.Y, expand=True)
#canvas.pack(side='top', fill=tkinter.BOTH, expand=True) # to expand vertical&horizontal use "BOTH, expand=True"
canvas.pack(side='left', anchor='n') # changed to left again to see the canvas oneside and buttons on other vertically

# fited the buttons on rightframe
rightframe = tkinter.Frame(mainwindow)
rightframe.pack(side='right', anchor='n', expand=True)


## to add the buttons
# fited the buttons on rightframe instead of mainwindow 
button1 = tkinter.Button(rightframe, text='button1')
button2 = tkinter.Button(rightframe, text='button2')
button3 = tkinter.Button(rightframe, text='button3')
#button1.pack(side='left') use anchor to fit the buttons at various  positions, and only wors for horizontal in this case, to 
## avoid that change pack's side to 'top' from 'left'on canvas.pack() and button.pack() and see the change.
## anchor uses 'n,w,s,e' and also compass points 'nw,se....'
button2.pack(side='top')
button3.pack(side='top')
button1.pack(side='top')

mainwindow.mainloop() ## this is complusory to run the window.
```

```python
""" GRID GEOMETRY MANAGER:..."""
## simpler than pack geometry and works by specifying the absolute positions for atleast one window.
## unless you know size of screen your runnig the progrm don't absolute psoitioning.
## for more complicated layouts or trivival layouts you get much better result using grid geometry.
## and unless you start adding things to it grid doesn't exists.
""" created a grid containing two rows 0 and 1 and 3 columns from 0 to 2 """

import tkinter

mainwindow = tkinter.Tk()

mainwindow.title('Hello world')
mainwindow.geometry('640x480-8-200') # -/+ to minimize or maximize the sizes of widgets of window.


## label, canvas, are widgets

label = tkinter.Label(mainwindow, text='Hello World')
label.grid(row=0, column=0) ## grid deals with rows and columns 

leftframe = tkinter.Frame(mainwindow)
leftframe.grid(row=1, column=1)

canvas = tkinter.Canvas(leftframe, relief='raised', borderwidth=1)
canvas.grid(row=1, column=0) 

# fited the buttons on rightframe
rightframe = tkinter.Frame(mainwindow)
rightframe.grid(row=1, column=2, sticky='n') ## sticky does same as anchor pionts in pack geometry


## to add the buttons

button1 = tkinter.Button(rightframe, text='button1')
button2 = tkinter.Button(rightframe, text='button2')
button3 = tkinter.Button(rightframe, text='button3')

button1.grid(row=0, column=0)
button2.grid(row=1, column=0)
button3.grid(row=2, column=0)

# configure the columns, means adding the weights to the columns.
mainwindow.columnconfigure(0, weight=1)
mainwindow.columnconfigure(1, weight=1)
mainwindow.grid_columnconfigure(2, weight=1) ## instead of column we used grid both are same(equal.)

leftframe.config(relief='sunken', borderwidth=1)
rightframe.config(relief='sunken', borderwidth=1)
leftframe.grid(sticky='ns')
rightframe.grid(sticky='new')


rightframe.columnconfigure(0, weight=1)
button2.grid(sticky='ew')

mainwindow.mainloop() ## this is complusory to run the window.
```

```python
""" ADVANCED GUI(TKINTER) EXAMPLE: """

import tkinter
import os

mainwindow = tkinter.Tk()

mainwindow.title('Grid Demo')
mainwindow.geometry('640x480-8-200')
mainwindow['padx'] = 8

label = tkinter.Label(mainwindow, text = 'Tkinter Grid Demo')
label.grid(row=0, column =0, columnspan=3)

mainwindow.columnconfigure(0, weight=100)
mainwindow.columnconfigure(1, weight=1)
mainwindow.columnconfigure(2, weight=1000)
mainwindow.columnconfigure(3, weight=600)
mainwindow.columnconfigure(4, weight=1000)
mainwindow.rowconfigure(0, weight=1)
mainwindow.rowconfigure(1, weight=10)
mainwindow.rowconfigure(2, weight=1)
mainwindow.rowconfigure(3, weight=3)
mainwindow.rowconfigure(4, weight=3)

## creating a listbox

filelist = tkinter.Listbox(mainwindow)
filelist.grid(row=1, column=0, sticky='nsew',rowspan=2)
filelist.config(borderwidth=1, relief='solid') # can solid, flat, grith, sunken etc....
for zone in os.listdir('/Windows/System32'):# '/user/bin' in MAC or LINUX. instead of typing and filling the list bax, using for loop and import OS
    filelist.insert(tkinter.END, zone) # when inserting each item into list using END constant will place each entry in the end of list as they are added, zero constant adds up reverse 

## creating a scroll bar for the list
listScroll = tkinter.Scrollbar(mainwindow, orient=tkinter.VERTICAL, command=filelist.yview) # if the scroll bar needs to be in horizontal, we use tkinter.HORIZONTAL, command=filelist.xview)
listScroll.grid(row=1,column=1,sticky='nsw',rowspan=2)
### linking the listbox with scroll bar.
filelist['yscrollcommand'] = listScroll.set

## frame for the radio buttons

optionframe = tkinter.LabelFrame(mainwindow, text='File Details') # LabelFrame is similiar to Frame but allows us to set caption for the frame and draws border around the context.
optionframe.grid(row=1, column=2, sticky='ne')

rbValue = tkinter.IntVar() # this line is for options on file details and using IntVar user can select any-one option which un-selects the previous option.
rbValue.set(3) # using set(can set defualt value 1/2/3...)
## radio button crearting

radio1 = tkinter.Radiobutton(optionframe, text='Filename',value=1,variable=rbValue)
radio2= tkinter.Radiobutton(optionframe, text='Path',value=2,variable=rbValue)
radio3= tkinter.Radiobutton(optionframe, text='TimeStamp',value=3,variable=rbValue)
radio1.grid(row=0, column=0, sticky='w')
radio2.grid(row=1,column=0,sticky='w')
radio3.grid(row=2,column=0,sticky='w')

## widget to display the result
resultlabel = tkinter.Label(mainwindow, text='Result')
resultlabel.grid(row=2, column=2, sticky = 'nw')
result = tkinter.Entry(mainwindow)
result.grid(row=2, column=2, sticky='sw')

# Frame for the timespinners

timeframe = tkinter.LabelFrame(mainwindow, text='Time')
timeframe.grid(row=3, column=0, sticky='new')

## time spinners
hourspinner = tkinter.Spinbox(timeframe, width=2, value=tuple(range(0,24))) ## two ways of adding values to spinners. below
minutespinner = tkinter.Spinbox(timeframe, width=2, from_=0, to=59) ## from_=0,to=59 gives 1to60, from is constant word so we use_
secondspinner = tkinter.Spinbox(timeframe, width=2, from_=0, to=59)
hourspinner.grid(row=0, column=0)
tkinter.Label(timeframe, text=':').grid(row=0, column=1)
minutespinner.grid(row=0, column=2)
tkinter.Label(timeframe, text=':').grid(row=0, column=3)
secondspinner.grid(row=0, column=4)
#kinter.Label(timeframe, text=':').grid(row=0, column=1)
timeframe['padx'] = 36 # put padding inside the left and right edges of the widget & 'pady' used to pad the top and bottom of the widget.

## Frame for date spinners

dateframe = tkinter.Frame(mainwindow)
dateframe.grid(row=4, column=0, sticky='new')

# date labels
daylabel = tkinter.Label(dateframe, text='Day')
monthlabel = tkinter.Label(dateframe, text='Month')
yearlabel = tkinter.Label(dateframe, text='Year')
daylabel.grid(row=0,column=0,sticky='w')
monthlabel.grid(row=0, column=1,sticky='w')
yearlabel.grid(row=0, column=2,sticky='w')
## date spinners

dayspin = tkinter.Spinbox(dateframe, width=5, from_=1, to=31)
monthspin = tkinter.Spinbox(dateframe, width=5, values=('Jan', 'Feb', 'Mar', 'Aprl', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'))
yearspin = tkinter.Spinbox(dateframe, width=5, from_=2000, to=2050)
dayspin.grid(row=1,column=0)
monthspin.grid(row=1, column=1)
yearspin.grid(row=1, column=2)

## buttons
okbutton = tkinter.Button(mainwindow, text='OK')
cancelbutton = tkinter.Button(mainwindow, text = 'Cancel', command=mainwindow.destroy) # can use .quit)
okbutton.grid(row=4, column=3, sticky='e')
cancelbutton.grid(row=4, column=4, sticky='w')


mainwindow.mainloop()

print(rbValue.get()) ## .get() here retrieves the value of the button were it is lastly selected.
```
