```python
## to check the ELAPSE TIME.

import time
from time import time as my_timer ## instead of time we can use monotonic, perf_counter, process_time
import random
## perf_counter- if we want to measure actual elapse time
##  process_time- if we want to measure how much time CPU spent on the task.
## time- to deal the real time rather than measuring use time.
input('Press ENTER to start')

wait_time = random.randint(1,6)
time.sleep(wait_time)
start_time = my_timer()
input('Please ENTER to stop')
end_time = my_timer()

print('STARTED AT:' + time.strftime('%X', time.localtime(start_time))) ## strftime()-string from time used to format local time tuple into more readable form
print('ENDED AT:' + time.strftime('%X', time.localtime(end_time)))
print('Your reaction time was {} seconds'.format(end_time - start_time))
```

```python
### checking the timezones
## 0 in time.gmtime(0)-represents the first possible start date of epoch

print('Epoch on this system starts at: '+ time.strftime('%c', time.gmtime(0)))
print('The current timezone is {0} with an off-set of {1}'.format(time.tzname[0], time.timezone))

if time.daylight != 0:
    print('\tDay light saving is in effect in this location')
    print('\tThe DST timezone' + time.tzname[1])
    
print('Local time is ' + time.strftime('%Y-%m-%d %H:%M:%S', time.localtime()))
print('UTC time is ' + time.strftime('%Y-%m-%d %H:%M:%S', time.gmtime()))
```

```python
## using python timezone(pytz) module, checking the timezone of countries

import pytz
import datetime

country = 'US/Alaska'

tz_to_display = pytz.timezone(country) ## using timezone(method)
local_time = datetime.datetime.now(tz=tz_to_display) ## using datetime.datetime.now(method)
print('The time in {} is  {}'.format(country, local_time))
print('The UTC time is {}'.format(datetime.datetime.utcnow()))
```

```python
# Create a program that allows a user to choose one of
# up to 9 time zones from a menu. You can choose any
# zones you want from the all_timezones list.
#
# The program will then display the time in that timezone, as
# well as local time and UTC time.
#
# Entering 0 as the choice will quit the program.
#
# Display the dates and times in a format suitable for the
# user of your program to understand, and include the
# timezone name when displaying the chosen time.


import pytz
import datetime

avail_zones = {'1': 'Asia/Yerevan',
               '2': 'Africa/Luanda',
               '3': 'Europe/Brussels',
               '4': 'Asia/Shanghai',
               '5': 'Asia/Kolkata',
               '6': 'Japan',
               '7': 'Asia/Baghdad',
               '8': 'Asia/Tokyo'
              }   # creating a dict with timezones

print('Please choose your timezone: ')
for place in sorted(avail_zones):  ## printing out the dict keys & values
    print('{}: {}'.format(place, avail_zones[place]))
    
while True:
    choice = input()   ## input from user
    
    if choice == '0':
        break   # breaks the loops and exits if user input is '0'.
        
    if choice in avail_zones.keys():
        tz_to_display = pytz.timezone(avail_zones[choice])
        world_time = datetime.datetime.now(tz=tz_to_display)
        print('The time in {} is {} {}'.format(avail_zones[choice], world_time.strftime('%A %x %X %z'), world_time.tzname()))
        print('Local time is {}'.format(datetime.datetime.now().strftime('%A %x %X'))) # our_location's time
        print('UTC time is {}'.format(datetime.datetime.utcnow().strftime('%A %x %X'))) # utc time
        print()
```
