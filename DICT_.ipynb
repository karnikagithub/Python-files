```python
"""Here's our solution to the challenge.

The code to add items to a dict is very similar to the original buy_computer program. That also used a dict.

Output will be a dict{}"""

avail_part = {"1": "computer",
                   "2": "monitor",
                   "3": "keyboard",
                   "4": "mouse",
                   "5": "hdmi cable",
                   "6": "dvd drive",
                   }
 
curr_choice = "-"
compt_part = {}  # create an empty dict
 
while curr_choice != "0":
    if curr_choice in avail_part:
        chosen_parts = avail_part[curr_choice]
        if curr_choice in compt_part:
            # it's already in, so remove it
            print(f"Removing {chosen_parts}")
            compt_part.pop(curr_choice)
        else:
            print(f"Adding {chosen_parts}")
            compt_part[curr_choice] = chosen_parts
        print(f"Your dictionary now contains: {compt_part}")
    else:
        print("Please add options from the list:")
        for key, value in avail_part.items():
            print(f"{key}: {value}")
        print("0: to finish")
 
    curr_choice = input("> ")
print('you\'ve chosen parts', compt_part)
```

```python
def deep_copy(d: dict) -> None:
    
    new_dic = {}
    for key, value in d.items():
        new_value = value.copy()
        new_dic[key] = new_value
        
    return new_dic

#recipes_copy = deep_copy(recipes)
print(recipes["butter_chicken"]['ginger']) = 300
print(recipes)
```
